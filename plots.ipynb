```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
a = [-100] + [10]*100
a = np.array(a)
```

```python
b = [np.npv(0.2, a[1:i+1]) for i in range(100)]
```

```python
fig, ax = plt.subplots()
for r in [0.02, 0.05, 0.1, 0.15, 0.2]:
    b = [np.npv(r, a[:i]) for i in range(101)]
    ax.plot(b[1:], label=r'$r={}$'.format(r))
ax.legend()
ax.set_xlabel('Facility life (years)', fontsize=12)
ax.set_ylabel('NPV', fontsize=12)
fig.tight_layout()
fig.savefig('npv_discount.svg')
plt.show()
```
